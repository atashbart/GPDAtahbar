```python
# Define xuv(x) function
def xuv(x, au, bu,cu,fu):
    return (2. *gamma(1. + au + bu) *gamma(1.5 + au + bu) *gamma(2. + au + bu))/(gamma(1 + bu)* (fu* gamma(1. + au) *gamma(1. + au + bu) *gamma(1.5 + au + bu) + cu *gamma(0.5 + au) *gamma(1. + au + bu) *gamma(2. + au + bu) + gamma(0. + au)* gamma(1.5 + au + bu) *gamma(2. + au + bu)))* np.power(x, au) * np.power(1 - x, bu)*(1+cu*np.power(x,0.5)+fu*x)
```

```python
# Define xdv(x) function
def xdv(x,ad, bd, cd,fd):
    return (gamma(1. + ad + bd) *gamma(1.5 + ad + bd)* gamma(2. + ad + bd))/(gamma(1 + bd) *(fd *gamma(1. + ad)* gamma(1. + ad + bd) *gamma(1.5 + ad + bd) + cd* gamma(0.5 + ad)* gamma(1. + ad + bd)* gamma(2. + ad + bd) + gamma(0. + ad)* gamma(1.5 + ad + bd)* gamma(2. + ad + bd)))* np.power(x, ad) * np.power(1 - x, bd)*(1+cd*np.power(x,0.5)+fd*x)
```

```python
def Hv(x, t, alpha, gamma, beta, m, be, ad, bd, cd,fd):
    # Ensure x is treated as a scalar inside the function
    return np.exp(alpha * t * (1 - x)**gamma * np.log(x) + beta * x**m * np.log(1 + be * t))*t*xdv(x, ad, bd, cd,fd)/x
```

```python
# Define F2(x) function
def f1pp(x, t,alpha,gamma,beta,m,be,au, bu, cu,fu, ad, bd, cd,fd):
    return (2.0/3.0* xuv(x, au, bu, cu,fu)/x - 1.0/3.0 * xdv(x, ad,bd, cd,fd)/x)*np.exp(alpha * t * (1 - x)**gamma * np.log(x) + beta * x**m * np.log(1 + be * t))*t
```

```python

def nd(ad, bd, cd,fd, etad):
    return (gamma (1. + ad + bd)*gamma (1.5 + ad + bd)*gamma (2. + ad + bd)*gamma (1 + bd + etad)*(gamma (ad)/gamma (1 + ad + bd + etad) + (1.*cd*gamma (0.5 + ad))/gamma (1.5 + ad + bd + etad) + (fd*gamma (1 + ad))/gamma (2 + ad + bd + etad)))/(gamma (1 + bd)* (fd*gamma (1. + ad)*gamma (1. + ad + bd)*gamma (1.5 + ad + bd) + (cd*gamma (0.5 + ad)*gamma (1. + ad + bd) +  gamma (0. + ad)*gamma (1.5 + ad + bd))*gamma (2. + ad + bd)))
```

```python
def  Hvv(x, t,alpha,gamma,beta,m,be,ad,bd, cd,fd,etad):
    return np.exp(alpha * t * (1 - x)**gamma * np.log(x) + beta * x**m * np.log(1 + be * t)) * t * xdv(x, ad, bd, cd,fd) / x * (2.033 / nd(ad, bd, cd,fd, etad)) * (1.0 - x)**etad
```

```python
def nu( au, bu, cu,fu,etau):
    return (2.*gamma (au + bu + 1.)*gamma (au + bu + 1.5)*gamma (au + bu + 2.)*gamma (bu + etau + 1)* ((1.*cu*gamma (au + 0.5))/ gamma (au + bu + etau + 1.5) + (fu*gamma (au + 1))/gamma (au + bu + etau +2) + (1.*gamma (au))/gamma (au + bu + etau + 1)))/ (gamma (bu + 1)*(gamma (au + bu + 2.)*(cu*gamma (au + 0.5)*gamma (au + bu + 1.) +  gamma (au + 0.)*gamma (au + bu + 1.5)) +   fu*gamma (au + 1.)*gamma (au + bu + 1.)* gamma (au + bu + 1.5)))
```
